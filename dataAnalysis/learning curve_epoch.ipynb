```python
file = 'learningcurve/03.sh.o62073'
def get_costs(file):
    with open(file, 'r') as f:
        valid_cost = []
        train_cost = []
        for line in f:
            if line.split()[0] == 'Epoch':
#                 print(line)
                continue
            elif line.split()[0] == 'Average':
#                 print(float(line.split()[2]))
                train_cost.append(float(line.split()[2]))
            elif line.split()[0] == 'Valid':
#                 print(float(line.split()[2]))
                valid_cost.append(float(line.split()[2]))
    return train_cost, valid_cost
```

```python
tc_char_sick, vc_char_sick = get_costs('learningcurve/03.sh.o62073')
tc_word_sick, vc_word_sick = get_costs('learningcurve/_04.sh.o62077')
tc_char_sts, vc_char_sts = get_costs('learningcurve/_07.sh.o62085')
tc_word_sts, vc_word_sts = get_costs('learningcurve/_08.sh.o62084')

print(len(tc_char_sick), len(vc_char_sick), len(tc_word_sick), len(vc_word_sick), len(tc_char_sts), len(vc_char_sts), len(tc_word_sts), len(vc_word_sts))
```
